```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
from sklearn.svm import LinearSVC

df = pd.read_csv('heart.csv')
df.head(303)
```

```python
feat1 = 'age'
feat2 = 'chol'
target = 'target'

plt.scatter(df[feat1], df[feat2], c=df[target])
plt.xlabel(feat1)
plt.ylabel(feat2)
plt.title("Age vs Chol")
plt.show()

X = df[[feat1, feat2]]
y = df[target]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
log_reg_model = LogisticRegression()
log_reg_model.fit(X_train, y_train)
y_pred = log_reg_model.predict(X_test)
accuracyScore = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1Score = f1_score(y_test, y_pred)
print("Logistic Regression Accuracy Score: ", accuracyScore)
print("Logistic Regression Precison Score: ", precision)
print("Logistic Regression Recall Score: ", recall)
print("Logistic Regression F1 Score: ", f1Score)

print('\n')

lin_svc_model = LinearSVC()
lin_svc_model.fit(X_train, y_train)
y_pred = lin_svc_model.predict(X_test)
accuracyScore = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1Score = f1_score(y_test, y_pred)
print("Linear SVC Accuracy Score: ", accuracyScore)
print("Linear SVC Precison Score: ", precision)
print("Linear SVC Recall Score: ", recall)
print("Linear SVC F1 Score: ", f1Score)

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
from sklearn.svm import LinearSVC, SVC

df = pd.read_csv('heart.csv')
df.head(303)
```

```python
numeric_features = ['age', 'cp', 'trestbps', 'chol', 'thalach', 'oldpeak', 'slope', 'ca', 'thal']

X = df[numeric_features]
y = df[target]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
log_reg_model = LogisticRegression()
log_reg_model.fit(X_train, y_train)
y_pred = log_reg_model.predict(X_test)
accuracyScore = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1Score = f1_score(y_test, y_pred)
print(f"Logistic Regression Accuracy Score: ", accuracyScore)
print(f"Logistic Regression Precison Score: ", precision)
print(f"Logistic Regression Recall Score: ", recall)
print(f"Logistic Regression F1 Score: ", f1Score)

print('\n')

lin_svc_model = LinearSVC()
lin_svc_model.fit(X_train, y_train)
y_pred = lin_svc_model.predict(X_test)
accuracyScore = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1Score = f1_score(y_test, y_pred)
print(f"Linear SVC Accuracy Score: ", accuracyScore)
print(f"Linear SVC Precison Score: ", precision)
print(f"Linear SVC Recall Score: ", recall)
print(f"Linear SVC F1 Score: ", f1Score)

print('\n')

kernel_svm_model = SVC()
kernel_svm_model.fit(X_train, y_train)
y_pred = kernel_svm_model.predict(X_test)
accuracyScore = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1Score = f1_score(y_test, y_pred)
print(f"Kernel SVM Accuracy Score: ", accuracyScore)
print(f"Kernel SVM Precison Score: ", precision)
print(f"Kernel SVM Recall Score: ", recall)
print(f"Kernel SVM F1 Score: ", f1Score)
```
